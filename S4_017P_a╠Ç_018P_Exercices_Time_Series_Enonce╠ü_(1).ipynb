# Exercice 2 : Entrainer un modèle

2.1 Reprendre les données françaises et entraîner un modèle Prophet. On récupérera 10% des données pour le jeu de test.

```python
from prophet import Prophet
```

```python
fig = px.line(df_france, x='date', y="nb_cas_covid") # Corrected line: Pass the entire DataFrame
fig.show()
```

```python
df_france.columns = ['ds', 'y']
df_france.head(3)
```

```python
df_france.shape
```

```python
train_size = df_france.shape[0] - (df_france.shape[0] * 0.1)
train_size = int(train_size)

train = df_france.iloc[:train_size, :]
test = df_france.iloc[train_size:, :]
print(train.shape, test.shape)
```

```python
def fit_prophet(df_train, df_test, freq, **kwargs):
  m = Prophet(seasonality_mode='multiplicative', changepoint_range=0.7)
  m.add_country_holidays('FR')

  m.fit(df_train)
  period = len(df_test)
  future = m.make_future_dataframe(periods=period, freq=freq)
  forecast = m.predict(future)

  return m, forecast
```

```python
m, forecast = fit_prophet(train, test, freq="d", seasonality_mode="multiplicative", changepoint_range=0.7)
```

2.2. Afficher les prédictions du modèle et les comparer aux données originales.

```python
future_forecast = forecast.iloc[-len(df_france):]
fig = px.line(future_forecast, x="ds", y="yhat", title ="True vs Predicted data")
fig.add_scatter(x=df_france["ds"], y=df_france["y"], mode="lines", name="True data")
fig.show()
```
